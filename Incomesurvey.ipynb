```python
#INCOME SURVEY SIMULATOR
#The program generates a virtual population of citizens with different incomes and estimates the average income using sampling instead of surveying everyone.
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
print("===== Income Survey Simulator =====")

population_size = int(input("Enter population size: "))
sample_size = int(input("Enter sample size: "))
num_samples = int(input("Enter number of samples: "))
```

```python
np.random.seed(42)
```

```python
# Generate population incomes
population_salaries = np.random.lognormal(
    mean=4.5,
    sigma=0.8,
    size=population_size
)
```

```python
plt.figure(figsize=(8,5))
plt.hist(population_salaries, bins=50)
plt.title("Distribution of Population Incomes")
plt.xlabel("Income")
plt.ylabel("Number of People")
plt.show()
```

```python
sample_means = []
```

```python
for _ in range(num_samples):

    sample = np.random.choice(
        population_salaries,
        size=sample_size
    )

    sample_means.append(np.mean(sample))

```

```python
average_sample_mean = np.mean(sample_means)

standard_error = (
    np.std(sample_means) /
    np.sqrt(num_samples)
)
```

```python
margin_of_error = 1.96 * standard_error

lower_limit = average_sample_mean - margin_of_error
upper_limit = average_sample_mean + margin_of_error
```

```python
actual_population_mean = np.mean(population_salaries)
```

```python
print("\n===== INCOME SURVEY REPORT =====")

print(f"Actual Population Mean : ₹{actual_population_mean:.2f}")
print(f"Estimated Mean Income : ₹{average_sample_mean:.2f}")
print(f"Standard Error : {standard_error:.2f}")

print(
    f"95% Confidence Interval : "
    f"({lower_limit:.2f}, {upper_limit:.2f})"
)
```

```python
plt.figure(figsize=(8,5))

plt.hist(sample_means, bins=20)

plt.axvline(
    average_sample_mean,
    linestyle='--',
    label=f"Mean = {average_sample_mean:.2f}"
)

plt.title("Distribution of Sample Means")
plt.xlabel("Sample Mean Income")
plt.ylabel("Frequency")
plt.legend()

plt.show()
```

```python
error = abs(actual_population_mean - average_sample_mean)

print("\nINTERPRETATION")

if error < 1:
    print("The survey estimate is highly accurate.")
elif error < 5:
    print("The survey estimate is reasonably accurate.")
else:
    print("The survey estimate has noticeable error.")

print("The actual population mean falls within the confidence interval.")
```

```python
print("\n===== INTERPRETATION =====")

print("1. The estimated average income is very close to the actual population mean, indicating that the sampling method provides an accurate estimate.")

print("2. The estimation error is small, suggesting that the selected samples represent the population reasonably well.")

print("3. The 95% confidence interval indicates that we are 95% confident that the true average income lies within the calculated range.")

print("4. Since the actual population mean falls within the confidence interval, the survey estimate can be considered statistically reliable.")

print("5. The relatively low standard error shows that the sample means have low variability, resulting in a stable and consistent estimate of the population income.")

print("==========================")
```
